```python
# Distribution graphs (histogram/bar graph) of column data
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    import numpy as np
    import matplotlib.pyplot as plt
    
    nCol = df.shape[1]
    nGraphRow = int(np.ceil(nGraphShown / nGraphPerRow))  # Ensure this is an integer
    
    plt.figure(num=None, figsize=(6 * nGraphPerRow, 8 * nGraphRow), dpi=80, facecolor='w', edgecolor='k')
    
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if not np.issubdtype(type(columnDf.iloc[0]), np.number):
            columnDf.value_counts().plot.bar()
        else:
            columnDf.hist()
        plt.title(columnDf.name)
    
    plt.tight_layout()
    plt.show()

```

```python
def plotCorrelationMatrix(df, graphWidth):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd
    
    filename = getattr(df, 'dataframeName', 'Dataset')  # Handle missing attribute for dataframe name
    
    # Filter only numeric columns
    numeric_df = df.select_dtypes(include=[np.number])
    
    # Ensure there are enough columns for a correlation matrix
    if numeric_df.shape[1] < 2:
        print(f"Correlation matrix plot requires at least 2 numeric columns, but {filename} has {numeric_df.shape[1]} column(s).")
        return
    
    corr = numeric_df.corr()  # Compute correlation matrix
    
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    plt.matshow(corr, fignum=1)
    plt.xticks(range(numeric_df.shape[1]), numeric_df.columns, rotation=90)
    plt.yticks(range(numeric_df.shape[1]), numeric_df.columns)
    plt.colorbar()
    plt.title(f"Correlation Matrix for {filename}", fontsize=15)
    plt.show()
```

```python
def plotScatterMatrix(df, plotSize, textSize):
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    from pandas.plotting import scatter_matrix
    
    df = df.select_dtypes(include=[np.number])  # keep only numerical columns
    # Remove rows and columns that would lead to df being singular
    df = df.dropna(axis=1)  # drop columns with NaN
    df = df[[col for col in df if df[col].nunique() > 1]]  # keep columns where there are more than 1 unique values
    columnNames = list(df)
    
    if len(columnNames) > 10:  # Limit number of columns to avoid overly complex scatter plot
        columnNames = columnNames[:10]
    
    df = df[columnNames]
    
    scatter_matrix(df, alpha=0.75, figsize=(plotSize, plotSize), diagonal='kde')
    corrs = df.corr().values
    
    for i, j in zip(*np.triu_indices_from(corrs, k=1)):
        plt.text(
            i, j, f"{corrs[i, j]:.2f}", 
            ha='center', va='center', 
            fontsize=textSize, 
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
        )
    plt.show()

```

```python
sdf = sklearn.utils.shuffle(df1)
print(sdf)
```

```python
data = sdf[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
train_data = data[0:100]
test_data = data[100:150]
labels = sdf['Species']
train_labels = labels[0:100]
test_labels = labels[100:150]
print(train_data, "\n", train_labels)
print(test_data,"\n", test_labels)
```

```python
from sklearn.neighbors import KNeighborsClassifier

scores = []
onetohundred = range(1,100)
for i in onetohundred:
    model = KNeighborsClassifier(n_neighbors=i)  # Choose k based on experimentation
    model.fit(train_data, train_labels)
    score = model.score(test_data, test_labels)
    scores.append(score)
plt.plot(onetohundred, scores)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train_data, train_labels)
print(model.score(test_data, test_labels))
```

```python
from sklearn.tree import DecisionTreeClassifier
scores = []
for i in onetohundred:
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(train_data, train_labels)
    scores.append(model.score(test_data, test_labels))
plt.plot(onetohundred, scores)
plt.show()

scores = []
onetotwenty = range(1,20)
for i in onetotwenty:
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(train_data, train_labels)
    scores.append(model.score(test_data, test_labels))
plt.plot(onetotwenty, scores)
plt.show()
print(scores)

print(DecisionTreeClassifier(max_depth=3).fit(train_data, train_labels).score(test_data, test_labels))
```

```python
from sklearn.linear_model import RidgeClassifier
numbers = np.array(range(1,300)) * 0.05
scores = []
for i in numbers:
    model = RidgeClassifier(alpha=i)
    model.fit(train_data, train_labels)
    scores.append(model.score(test_data, test_labels))

plt.plot(numbers, scores)
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier

numbers = np.array(range(1,100)) * 4
scores = []
for i in numbers:
    model = ExtraTreesClassifier(n_estimators=i)
    model.fit(train_data, train_labels)
    scores.append(model.score(test_data, test_labels))

plt.plot(numbers, scores)
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

numbers = np.array(range(1,30))
scores = []
for i in numbers:
    base_model = DecisionTreeClassifier(max_depth=3)
    model = BaggingClassifier(base_model, n_estimators=i)
    model.fit(train_data, train_labels)
    scores.append(model.score(test_data, test_labels))
    #changes everytime ran

plt.plot(numbers, scores)
plt.show()
```

```python
from sklearn.neighbors import NearestCentroid

model = NearestCentroid()
model.fit(train_data, train_labels)
print(model.score(test_data, test_labels))
```

```python
from sklearn.gaussian_process import GaussianProcessClassifier
model = GaussianProcessClassifier()
model.fit(train_data, train_labels)
print(model.score(test_data, test_labels))
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split

df = pd.read_csv('input/Iris.csv')

features = df.drop(columns=['Id', 'Species']) 
labels = df['Species']

encoder = LabelEncoder()
labels = encoder.fit_transform(labels)

train_data, test_data, train_labels, test_labels = train_test_split(features, labels, test_size=0.3, random_state=32)

scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)

# Define the model
nclasses = len(np.unique(train_labels)) 
model = Sequential([
    Input(shape=(train_data.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(nclasses, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_data, train_labels, epochs=50, batch_size=8)
model.evaluate(test_data, test_labels)
```
